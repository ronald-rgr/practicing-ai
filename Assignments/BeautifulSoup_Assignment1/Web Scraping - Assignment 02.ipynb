#### 1: View and print the Simplilearn webpage content in a good format

```python
#Import the required libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
#Provide the url for web scraping
url ='file:///Users/ronald.garciarobles/Ron/Architecture/Artificial%20Intelligence/AI%20Engineer/Assignments/Assignment1_BeautifulSoup/BC%20Bid_Construction.html'
```

```python
#Scrape the web page document
webpage = urlopen(url)
```

```python
#Create the soup object and parse the webpage using an html parser
sl_soup = BeautifulSoup(webpage,"html.parser")
```

```python
#Close the webpage object
webpage.close()
```

```python
#View the soup contents
sl_soup.contents
```

```python
#Use the prettify() method to view the well-formatted html output
print(sl_soup.prettify())
```
